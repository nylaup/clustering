**Q1.** This is a question about clustering. We want to investigate how adjusting the "noisiness" of the data impacts the quality of the algorithm and the difficulty of picking $k$.

1. Run the code below, which creates four datasets: `df0_125`, `df0_25`, `df0_5`, `df1_0`, and `df2_0`. Each data set is created by increasing the amount of `noise` (standard deviation) around the cluster centers, from `0.125` to `0.25` to `0.5` to `1.0` to `2.0`.


````python
import numpy as np
import pandas as pd

def createData(noise,N=50):
    np.random.seed(100) # Set the seed for replicability
    # Generate (x1,x2,g) triples:
    X1 = np.array([np.random.normal(1,noise,N),np.random.normal(1,noise,N)])
    X2 = np.array([np.random.normal(3,noise,N),np.random.normal(2,noise,N)])
    X3 = np.array([np.random.normal(5,noise,N),np.random.normal(3,noise,N)])
    # Concatenate into one data frame
    gdf1 = pd.DataFrame({'x1':X1[0,:],'x2':X1[1,:],'group':'a'})
    gdf2 = pd.DataFrame({'x1':X2[0,:],'x2':X2[1,:],'group':'b'})
    gdf3 = pd.DataFrame({'x1':X3[0,:],'x2':X3[1,:],'group':'c'})
    df = pd.concat([gdf1,gdf2,gdf3],axis=0)
    return df

df0_125 = createData(0.125)
df0_25 = createData(0.25)
df0_5 = createData(0.5)
df1_0 = createData(1.0)
df2_0 = createData(2.0)
````

2. Make scatterplots of the $(X1,X2)$ points by group for each of the datasets. As the `noise` goes up from 0.125 to 2.0, what happens to the visual distinctness of the clusters?    
As you increase the noise, there is less distinctness of the clusters and X1 and X2 are much less distinct. At noise of 0.125 you can see the distinction of X1 and X2 at 5 and 2, even though they are mixed together at 1 and 3. However, at a noise level of 2 there is almost no distinct clustering and the points are all scattered together.  

````python
import matplotlib.pyplot as plt 
import seaborn as sns

sns.scatterplot(df0_125)
plt.show()
sns.scatterplot(df0_25)
plt.show()
sns.scatterplot(df0_5)
plt.show()
sns.scatterplot(df1_0)
plt.show()
sns.scatterplot(df2_0)
plt.show()
````

3. Create a scree plot for each of the datasets. Describe how the level of `noise` affects the scree plot (particularly the presence of a clear "elbow") and your ability to definitively select a $k$. (Pay attention to the vertical axis across plots, or put all the scree curves on a single canvas.)
As you increase the noise, it looks like the elbow point has a higher SSE. The graphs are also much less drastic and so it is easier to find a clear elbow when there is lower noise, which will make it easier to definitively select a k. 

````python
from sklearn.cluster import KMeans

#create a dictionary of all of the datasets to be able to loop through them
df_dict = {"df0_125": df0_125, "df0_25": df0_25, "df0_5": df0_5, "df1_0": df1_0, "df2_0": df2_0}

def maxmin(x): 
    x = (x-min(x))/(max(x)-min(x))
    return x


def find_sse(df):
    X = df.loc[:,['x1','x2']].apply(maxmin) #normalize all datasets
    k_bar = 10
    SSE = []
    for k in range(k_bar):
        model = KMeans(n_clusters=k+1, max_iter=300, n_init = 10, random_state=0) 
        model = model.fit(X) 
        SSE.append(model.inertia_) 
    return SSE 

SSE_results = {}
for name,df in df_dict.items(): 
    SSE_results[name] = find_sse(df) #create a dictionary of all of the SSE results for each df 

k_bar = 10
for name, SSE in SSE_results.items():
    sns.lineplot(x=np.arange(1,k_bar+1), y=SSE, label=name) #plot each df and its SSE results
plt.title('Scree Plot for all dfs')
plt.legend()
plt.show()
````

3. Explain the intuition of the elbow, using this numerical simulation as an example.
The elbow is the point in the plot where the SSE stops decreasing drastically, and this is the k value you should choose for kmeans. For df0_125 this is around 3, and after that the SSE is almost horizontal. When creating kmeans you are trying to achieve the lowest SSE without creating too many clusters. This is why you choose the point where there is an elbow point, because if you did any k value before the SSE would be too high, while if you did any point after there would be too many clusters. For it to be computationally optimal, you choose a k value at the elbow point.
